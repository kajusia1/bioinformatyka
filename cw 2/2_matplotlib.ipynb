## 2. Generowanie danych do wykresu
<font size=2>Podążaj za wskazówkami w opisach i komentarzach umieszczonych w komórkach, celem wykonania prostego wykresu liniowego. Jako pomoc wykorzystaj informacje zawarte w prezentacji z wykładu.
    
Zapoznaj się z funkcjami, z którymi będziesz pracował(a):
    

| Rodzaj funkcji | |
|-|-|
| kwadratowa | $y(x) = 0.5 \cdot x^{2} - 2.5 \cdot x + 0.5$ |
| sinusoidalna | $y(x) = 8.5 \cdot sin(x)$ |
| oraz ich suma | $y(x) = 0.5 \cdot x^{2} - 2.5 \cdot x + 0.5 + 8.5 \cdot sin(x)$ |

</font>

```python
# uruchom poniższy kod,
# aby Jupyter nie zmieniał marginesów wykresu,
# którego elementy znajdują się poza grafiką,
# tak aby wykres w notatniku wyglądał
# dokładnie jak ten zapisany do pliku

%config InlineBackend.print_figure_kwargs = {'bbox_inches':None}
```

```python
# zaimportuj moduł numpy pod nazwą 'np'
# oraz matplotlib.pyplot pod nazwą 'plt'

import numpy as np
import matplotlib.pyplot as plt
```

```python
# w tej komórce:
# - utwórz wektor X wartości od -10.0 do 15.0 co 0.2
# - utwórz wektor Y_square będący wektorem wartości funkcji kwadratowej
# - utwórz wektor wartości Y_sin będący wektorem wartości funkcji sinusoidalnej,
#   skorzystaj z funkcji np.sin(...)
# - utwórz wektor Y będący sumą Y_square i Y_sin
X = np.arange(-10.0, 15.0, 0.2)

Y_square = 0.5*X**2 -2.5*X + 0.5

Y_sin = 8.5 *np.sin(X)

Y=Y_square+ Y_sin

```

---

## 3. Prosty wykres liniowy

<font size=2>
    
Narysuj wykres `Y(X)`, całkowicie podstawowy, o rozmiarze 5x4 cale i rozdzielczości 150 dpi. Następnie zapisz go do pliku `output/chart_1.png` z nieprzezroczystym tłem i w rozdzielczości 300 dpi.
    
</font>

```python

plt.figure(figsize=(5, 4), dpi=150)  # Ustawienie rozmiaru 5x4 cale i rozdzielczości 150 dpi
plt.plot(X, Y, label='Y(X)')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()

# Zapis do pliku z rozdzielczością 300 dpi i nieprzezroczystym tłem
plt.savefig('output/chart_1.png', dpi=300, transparent=False)

# Wyświetlanie wykresu
plt.show()
```

## 4. Formatowanie wykresu liniowego

<font size=2>Opis metod właściwych dla obiektu osie (Axes, `ax`), z których będziesz musiał(a) skorzystać, znajdziesz [tutaj](https://matplotlib.org/3.1.1/api/axes_api.html#axis-limits).
\
\
Skopiuj kod z wcześniejszej komórki oraz <u>usuń linię kodu rysującą wykres liniowy</u>.
\
\
W obrębie utworzonej figury i na tych samych osiach narysuj kolejno trzy różne wykresy:
    
1. czerwoną (parametr `color`) linią o grubości (`linewidth`) 0.8 punktu i przezroczystości (`alpha`) równej 0.5, narysuj wykres `Y_square(X)`;
1. niebieską linią o grubości 0.8 punktu i przezroczystości równej 0.5, narysuj wykres `Y_sin(X)`;
1. wykres `Y(X)` narysuj w postaci okrągłych punktów (`marker='o'`) o rozmiarze (`markersize`) 3 punkty, o kolorze wypełnienia (`markerfacecolor`) czerwonym, kolorze obwódki (`markeredgecolor`) niebieskim, grubości obwódki (`markeredgewidth`) równej 0.5 punktu, bez linii (`linestyle=' '`)

Dodaj legendę do wykresu w postaci listy etykiet o wartościach `Square, Sin, Sum`, korzystając z metody `ax.legend(...)`.

Następnie dodaj do osi tytuł wykresu oraz etykiety osi (np. x i y), wykorzystując metody `ax.set_title(...), ax.set_xlabel(...), ax.set_ylabel(...)`. Jako tytuł ustaw pełny, matematyczny wzór funkcji `Y(X)`, posługując się wyrażeniem $\LaTeX$. Poza informacjami zawartymi w notatniku nr 0, pomocnym tutaj będzie jeszcze wiedzieć, że w wyrażeniach $\LaTeX$ kropka oznaczająca mnożenie to `\cdot`.
\
\
Jeżeli wykres nie mieści się w obrębie grafiki, dostosuj położenie osi w wywołaniu metody `fig.add_axes(...)`.
\
\
Wykres zapisz do pliku `output/chart_2.png`.
\
\
</font>

```python
fig = plt.figure(figsize=(5, 4), dpi=150)
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])  

ax.plot(X, Y_square, color='red', linewidth=0.8, alpha=0.5, label='Square')
ax.plot(X, Y_sin, color='blue', linewidth=0.8, alpha=0.5, label='Sin')
ax.plot(
    X, Y, marker='o', markersize=3, markerfacecolor='red', 
    markeredgecolor='blue', markeredgewidth=0.5, linestyle=' ', 
    label='Sum'
)
ax.legend(['Square', 'Sin', 'Sum'])
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()

# Tytuł i etykiety osi
ax.set_title(r"$Y(X) = X^2 + \sin(X)$")  # LaTeX title for the function
ax.set_xlabel('X')
ax.set_ylabel('Y')

# Zapis do pliku z rozdzielczością 300 dpi
plt.savefig('output/chart_2.png', dpi=300, transparent=False)
```

<font size=2>
    
Na koniec utwórz nową komórkę, skopiuj do niej działający kod z poprzedniej celem przetestowania różnych opcji formatowania wykresu liniowego. Dokumentacja metody `ax.plot(...)` znajduje się [tutaj](https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.plot.html). Przetestuj w szczególności opcje formatowania osi, np. zakresy wartości (`ax.set_xlim(...)` i `ax.set_ylim(...)`) czy gęstość podziałki (`ax.set_xticks(...)` i `ax.set_yticks(...)`), czy też etykiety znaczników podziałki (`ax.set_xticklabels(...)` i `ax.set_yticklabels(...)`).
</font>

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(5, 4), dpi=150)
ax = fig.add_axes([0.15, 0.15, 0.8, 0.8])

plt.plot(X,Y_square, color='red',linewidth=0.8, alpha=0.5)
plt.plot(X,Y_sin, color='blue',linewidth=0.8, alpha=0.5)
plt.plot(X,Y, marker='o', markersize=3, markerfacecolor='yellow', markeredgecolor='green', markeredgewidth=0.5, linestyle=' ')
ax.legend(['Square', 'Sin', 'Sum'])
ax.set_title(r'$y(x) = 8.5 \cdot \sin(x) + 0.5 \cdot x^2 - 2.5 \cdot x + 0.5$')
ax.set_xlabel('Oś x')
ax.set_ylabel('Oś y')
ax.set_xlim(-10,10)
ax.set_ylim(-20,120)
ax.set_xticks([1, 3, 21, 23, 25, 28])
ax.set_yticks([10, 15, 20])
ax.set_xticklabels(['c', 'o', 't', 'a', 'm', '?'])
ax.set_yticklabels(['j','e', 'h'])
#fig.savefig('output2/chart_2.png', transparent=False, dpi=300)
```
